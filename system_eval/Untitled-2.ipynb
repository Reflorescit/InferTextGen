```python
def read_jsonl(filename):
    data = []
    with open(filename, "r") as f:
        for line in f:
            data.append(json.loads(line))
    return data

data = read_jsonl('./gpt2_zeroshot/gpt2-xl-zeroshot-atomic2020.jsonl')
```
